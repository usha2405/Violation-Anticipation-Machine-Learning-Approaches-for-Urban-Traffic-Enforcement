```python
# convert time to 24-hour format
def convert_time(t):
    if pd.isnull(t) or not isinstance(t, str) or len(t) != 5 or t[-1] not in ['A', 'P']:
        return None, None  # Return None for both hour and minute for invalid times
    hour, minute, period = int(t[:2]), int(t[2:4]), t[-1]
    if period == 'P' and hour < 12:
        hour += 12
    if period == 'A' and hour == 12:
        hour = 0
    hour = f"{hour:02d}"
    if int(hour)==43:
      hour='4'
    elif int(hour)==50:
      hour='5'
    return int(hour), int(f"{minute:02d}")
# Apply conversion to 'Violation Time' and split into two new columns
df['Violation Hour'], df['Violation Minute'] = zip(*df['Violation Time'].apply(convert_time))

df['Violation Hour'] = df['Violation Hour'].fillna(0)
df['Violation Minute'] = df['Violation Minute'].fillna(0)

df.head()
```

#Data Transformation: Binning

```python
def timeOfTheDay(h):
  hour=int(h)
  if 00 <= hour < 6:
    return 0
  elif 6 <= hour < 12:
    return 1
  elif 12 <= hour or hour < 18:
    return 2
  else:
    return 3


def hourOfTheDay(h):
  hour=int(h)
  if 0 <= hour < 15:
    return 0
  elif 15 <= hour < 30:
    return 1
  elif 30 <= hour < 45:
    return 2
  elif 45 <= hour or hour < 60:
    return 3


def vehicle_yr(y):
  year=int(y)
  if 1970<year<1980:
    return 1
  elif year<1990:
    return 2
  elif year<2000:
    return 3
  elif year<2010:
    return 4
  elif year<2020:
    return 5
  elif year<=2030:
    return 6
  else:
    return 0

df['Time of Day'] = df['Violation Hour'].apply(timeOfTheDay)
df['Hour of Day'] = df['Violation Minute'].apply(hourOfTheDay)
df['Vehicle Year Binned'] = df['Vehicle Year'].apply(vehicle_yr)
```

#Threshholding

```python
import pandas as pd
import numpy as np

# Assuming df is your initial DataFrame
initial_count = len(df)
print("Initial count:", initial_count)

# Function to apply a quantile-based filter on a column
def filter_df_on_quantile(df, column_name, quantile=0.90):
    # Calculate the 90th percentile threshold from value counts
    threshold = df[column_name].value_counts().quantile(quantile)
    # Filter and return DataFrame where column values meet the threshold
    df_filtered = df[df[column_name].isin(df[column_name].value_counts()[df[column_name].value_counts() >= threshold].index)]
    return df_filtered, len(df_filtered)

# List of columns to apply the filter on, in sequence
columns = ['Registration State', 'Plate Type',
    'Vehicle Body Type', 'Vehicle Make',  'Vehicle Expiration Date',
 'Issuer Code',   'Date First Observed',
     'Feet From Curb', 'Issue Date year'
]

filter_counts = {}  # Dictionary to store counts after each filter

for column in columns:
    df, count = filter_df_on_quantile(df, column)
    filter_counts[column] = count  # Store count associated with each column
    print(f"Count after filtering on {column}: {count}")

```
